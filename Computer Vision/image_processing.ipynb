```python
import cv2
cv2.__version__
```

```python
import numpy as np
import matplotlib.pyplot as plt

#import image
img = cv2.imread('image.jpeg')
plt.imshow(img)
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray, cmap='gray')
```

```python
# RGB images are made up of 3 channels: R, G, and B
fig, axs = plt.subplots(nrows=1, ncols=4, figsize=(20,20))
plt.imshow(img)

for i in range(0,3):
    ax = axs[i]
    ax.imshow(img[:,:,i], cmap='gray')

plt.show()
# in each channel from (left to right is RGB)
# higher concentration of blue in an original image is whiter in blue channel
```

```python
# Transform the image into HSV (hue, saturation, and value) and HLS (hue, saturation, and lightness) models
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
img_hls = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
# Plot the converted images
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(20,20))
ax1.imshow(img_hsv)
ax2.imshow(img_hls)
plt.show()
# has a variety of usecases, such as lane detection
```

```python
# DRAWING ON IMAGES
img_copy = img.copy()

# draw a rectangle
cv2.rectangle(img_copy, pt1 = (10,250), pt2 = (150,400), color=(255,0,0),thickness=5)
cv2.circle(img_copy, center=(365, 555), radius=100, color=(0,0,255), thickness=5)
cv2.putText(img_copy, text="unreal", org = (250, 250),fontFace = cv2.FONT_HERSHEY_DUPLEX, fontScale=2, color=(0,255,0), thickness=3, lineType=cv2.LINE_AA)
plt.imshow(img_copy)
```
